```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as knc
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/phpXeun7q.csv")
```

```python
dtst
```

```python
train,test=train_test_split(dtst,test_size=0.2)
trainx=train.iloc[:,0:20]
trainy=train.iloc[:,20]
testx=test.iloc[:,0:20]
testy=test.iloc[:,20]
```

```python
trainx
```

```python
trainy
```

```python
ac=[]
for i in range(3,100,2):
    mdlkn=knc(n_neighbors=i)
    mdlkn.fit(trainx,trainy)
    prdkn=mdlkn.predict(testx)
    knac=np.mean(prdkn==testy)
    ac.append([knac,i])
```
