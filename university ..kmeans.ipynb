```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/Universities.csv")
dtst
```

```python
x=dtst.iloc[:,1:6]
x
```

```python
kmean = KMeans(n_clusters=5)
y_kmeans5 = kmean.fit_predict(x)
print(y_kmeans5)
kmean.cluster_centers_
university =[]
```

```python
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    university.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 11), university)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('outcome')
plt.show()
```

```python
kmeans3 = KMeans(n_clusters=3)
y_kmeans3 = kmeans3.fit_predict(x)
print(y_kmeans3)
```
