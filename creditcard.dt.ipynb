```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/Credit_Card_Applications (1).csv")
```

```python
dtst
```

```python
train,test=train_test_split(dtst,test_size=0.3)
trainx=train.iloc[:,0:15]
trainy=train.iloc[:,15]
testx=test.iloc[:,0:15]
testy=test.iloc[:,15]
```

```python
trainx
```

```python
trainy
```

```python
mdldt=DecisionTreeClassifier(criterion="entropy")
mdldt.fit(trainx,trainy)
preddt=mdldt.predict(testx)
cm1=confusion_matrix(preddt,testy)
acc=np.mean(preddt==testy)
```
