```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as knc
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/Multi Linear Regression.csv")
```

```python
dtst
```

```python
dtst.isnull().sum()
```

```python
train,test=train_test_split(dtst,test_size=0.3)
trainx=train.iloc[:,0:13]
trainy=train.iloc[:,13]
testx=test.iloc[:,0:13]
testy=test.iloc[:,13]
```

```python
ac=[]
for i in range(3,100,2):
    mdl=knc(n_neighbors=i)
    mdl.fit(trainx,trainy)
    prd=mdl.predict(testx)
    acc=np.mean(prd==testy)
    ac.append([acc,i])
```
