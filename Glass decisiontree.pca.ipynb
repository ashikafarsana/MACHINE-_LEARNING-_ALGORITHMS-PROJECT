```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/glass.csv")
```

```python
target=dtst.iloc[:,9]
target
```

```python
features=dtst.iloc[:,0:9]
features
```

```python
scalar=StandardScaler()
scalar.fit(features)
dt=scalar.transform(features)
```

```python
dt
```

```python
pca=PCA(n_components=2)
pca.fit(dt)
cv=pca.transform(dt)
cv
```

```python
cvv=pd.DataFrame(cv)
```

```python
cvv
```

```python
X_train, X_test, y_train, y_test = train_test_split(cvv,target, test_size = 0.3, random_state=1)
```

```python
mdl=DecisionTreeClassifier(criterion='entropy')
mdl.fit(X_train,y_train)
prdict=mdl.predict(X_test)
acc=np.mean(prdict==y_test)
acc
```
