```python
from apyori import apriori
```

```python
data=pd.read_csv(r"C:\Users\hp\Downloads\Market_Basket_Optimisation.csv",header=None)
data.head()
```

```python
data.shape
```

```python
#converting dataframe into a list of lists
l=[]
for i in range(1,7501):
    l.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
#Applying Apriori Algorithm
ar=apriori(l,min_support=0.0045,min_confidence=0.2,min_lift=3,min_length=2)
a_results=list(ar)
```

```python
for i in range(0,len(a_results)):
    print(a_results[i][0])
```

```python
for item in a_results:
    pair=item[0]
    items=[x for x in pair]
    print("Rule:"+items[0]+"-->"+items[1])
    print("Support:"+str(item[1]))
    print("Confidence:"+str(item[2][0][2]))
    print("Lift:"+str(item[2][0][3]))
    print("---------------------------------------------------------")
```
