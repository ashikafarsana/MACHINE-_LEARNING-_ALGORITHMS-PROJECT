```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier as knc
from sklearn.model_selection import train_test_split
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/Credit_Card_Applications (1).csv")
```

```python
dtst
```

```python
train,test=train_test_split(dtst,test_size=0.3)
trainx=train.iloc[:,0:15]
trainy=train.iloc[:,15]
testx=test.iloc[:,0:15]
testy=test.iloc[:,15]
```

```python
trainx
```

```python
trainy
```

```python
c=[]
for i in range(3,100,2):
    mdl=knc(n_neighbors=i)
    mdl.fit(trainx,trainy)
    prd=mdl.predict(testx)
    acc=np.mean(prd==testy)
    c.append([acc,i])
```
