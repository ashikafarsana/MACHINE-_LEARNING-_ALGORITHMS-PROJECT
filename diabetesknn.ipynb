```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNC
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/diabetes.csv")
```

```python
train,test=train_test_split(dtst,test_size=0.2)
trainx=train.iloc[:,0:8]
trainy=train.iloc[:,8]
testx=test.iloc[:,0:8]
testy=test.iloc[:,8]
```

```python
trainx
```

```python
trainy
```

```python
hs=[]
for i in range(3,50,2):
    modelkn=KNC(n_neighbors=i)
    modelkn.fit(trainx,trainy)
    prdkn=modelkn.predict(testx)
    knac=np.mean(prdkn==testy)
    hs.append([knac,i])
```
