```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/concrete_data.csv")
```

```python
dtst
```

```python
dtst.to_string()
```

```python
kme= KMeans(n_clusters=5)
model=kme.fit_predict(dtst)
print(model)
kme.cluster_centers_
stri=[]
```

```python
for i in range(1,11):
    km=KMeans(n_clusters=i).fit(dtst)
    km.fit(dtst)
    stri.append(km.inertia_)
    
```

```python
stri
```

```python
plt.plot(range(1, 11),stri)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('outcome')
plt.show()
```
