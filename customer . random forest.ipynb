```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/Multi Linear Regression.csv")
```

```python
dtst
```

```python
train,test=train_test_split(dtst,test_size=0.3)
trainx=train.iloc[:,0:13]
trainy=train.iloc[:,13]
testx=test.iloc[:,0:13]
testy=test.iloc[:,13]
```

```python
trainx
```

```python
testy
```

```python
modelrf=RandomForestClassifier(n_jobs=13,oob_score=True,n_estimators=100,criterion="entropy")
modelrf.fit(trainx,trainy)
prd=modelrf.predict(testx)
cm=confusion_matrix(testy,prd)
```
