```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
dtst=pd.read_csv("C:/Users/PARCHU/Downloads/Credit_Card_Applications (1).csv")
```

```python
dtst
```

```python
train,test=train_test_split(dtst,test_size=0.3)
trainx=train.iloc[:,0:15]
trainy=train.iloc[:,15]
testx=test.iloc[:,0:15]
testy=test.iloc[:,15]
```

```python
trainx
```

```python
trainy
```

```python
gnb=GaussianNB()
mnb=MultinomialNB()
bnb=BernoulliNB()
```

```python
pred_gnb=gnb.fit(trainx,trainy).predict(testx)
pred_mnb=mnb.fit(trainx,trainy).predict(testx)
pred_bnb=bnb.fit(trainx,trainy).predict(testx)
cm1=confusion_matrix(testy,pred_gnb)
accgnb=np.mean(pred_gnb==testy.values.flatten())
cm2=confusion_matrix(testy,pred_mnb)
accmnb=np.mean(pred_mnb==testy.values.flatten())
cm3=confusion_matrix(testy,pred_mnb)
accbnb=np.mean(pred_bnb==testy.values.flatten())

```
