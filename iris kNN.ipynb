```python
import pandas as p
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNC
```

```python
iris=p.read_csv("C:/Users/hp/Downloads/iris.csv")
```

```python
iris
```

```python
train,test=train_test_split(iris,test_size=0.2)
trainx=train.iloc[:,0:4]
trainy=train.iloc[:,4]
testx=test.iloc[:,0:4]
testy=test.iloc[:,4]
```

```python
trainx
```

```python
trainy
```

```python
c=[]
for i in range(3,100,2):
    mdl=KNC(n_neighbors=i)
    mdl.fit(trainx,trainy)
    prd=mdl.predict(testx)
    acc=np.mean(prd==testy)
    c.append([acc,i])
```
